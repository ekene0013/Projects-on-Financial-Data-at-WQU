## Mortgage 1
**30 Year Fixed Rate**

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 500)
```

```python
total = 1000000
rate = 0.04/12
# We create a list of month index
month=list(range(1,361))

# Calculate the discount fator for each month
c = [1/(1+rate)**i for i in month]
# The total payment should equal sum of discounted all future cash flow.
payment = total/sum(c)

principal_paid = 0

# We create an empyt list here, and append the result for each month.
# Lists are useful here because of its dynamic nature, which means that they can be modified after they are created by adding or removing elements.
# For example, you can add an element to a list using the append() method.
data = []
for m in month:
    interest = (total - principal_paid)*rate
    principal = payment - interest
    principal_paid += principal
    new_principal_balance = total - principal_paid
    data.append([m, payment, interest, principal, new_principal_balance] )

# month, payment, interest, principal, new_principal_balance
data[:5]
```

For easy review, we transform the list we get into a DataFrame.

```python
pd.DataFrame(data,columns = [ 'month', 'payment', 'interest', 'principal', 'new_principal_balance'])
```

```python
# Calculate payment per month helper function
def helper1 (years, new_principal_balance, interest):
    top = new_principal_balance * (interest /12) * (1 + (interest / 12))**(years)
    bottom = (1 + (interest/12))**(years) - 1
    return top/ bottom
```
